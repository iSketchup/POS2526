# SQL- Data definiton Language (DDL)
Tabellen erstellen und die Struktur ändern wie z.B : Attribute ändern/löschen/neu ... 

#### Befehle 
- CREATE TABLE
- ALTER TABLE
- DROP TABLE

#### Datentypen 
Siehe ERM skript Nummer 39

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
cur = conn.cursor()

cur.execute("""
CREATE TABLE schueler(
            schuelerID INTEGER PRIMARY KEY,
            name VARCHAR(50) NOT NULL,
            "alter" INTEGER NOT NULL);
            """)

comand = "PRAGMA TABLE_INFO('schueler');"

print(cur.execute(comand).fetchall())

pd.read_sql(comand, conn)

cur.execute("""ALTER TABLE schueler 
            ADD klasse VARCHAR(10) NOT NULL;
            """)

# cur.execute("""DROP TABLE schueler;""")

cur.execute("""
INSERT INTO schueler (name, "alter", klasse) VALUES 
            ('tim', 16, '4a'),
            ('lisa', 17, '4b'),
            ('anna', 16, '4a'),
            ('paul', 15, '3c');
""")


cur.execute("SELECT * FROM schueler").fetchall()

pd.read_sql("SELECT * FROM schueler", conn)


cur.execute("""ALTER TABLE schueler 
            RENAME COLUMN 'alter' TO age;
            """)


cur.execute("""ALTER TABLE schueler 
            DROP COLUMN klasse;
            """)

pd.read_sql("SELECT * FROM schueler", conn)

```

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect(":memory:")
cur = conn.cursor()


cur.execute("""
CREATE TABLE people(
            ID INTEGER PRIMARY KEY,
            HANDLE TEXT,
            NAME TEXT,
            AGE INTEGER,
            BALANCE INTEGER,
            IS_ADMIN BOOLEAN);
            
            """)

cur.execute("""ALTER TABLE people 
            RENAME to users;
            """)
           
cur.execute("""ALTER TABLE users  
            RENAME COLUMN HANDLE TO username;
            """)

cur.execute("ALTER TABLE users ADD COLUMN passsword VARCHAR(50);")

cur.execute("ALTER TABLE users DROP COLUMN IS_ADMIN;")

pd.read_sql("PRAGMA TABLE_INFO('users');", conn)
```
