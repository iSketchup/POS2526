# 2. Data Manipulation Language (DML)

Um Daten in einer Tabelle ändern/einfügen oder löschen können.

- INSERT INTO: Fügt Daten einer Tabelle hinzu
- UPDATE: Aktualisiert Daten einer
- DELETE: Löscht Daten einer Tabelle



### Beispiel
1. Tabelle Schueler soll gelöscht werden, damit wir immer eine frische Tabelle zum Arbeiten haben.
2. Tabelle Schueler soll mit folgenden Attributen erstellt werden:

|   cid | name   | type        |   notnull | dflt_value   |   pk |
|------:|:-------|:------------|----------:|:-------------|-----:|
|     0 | id     | INTEGER     |         0 |              |    1 |
|     1 | name   | VARCHAR(50) |         0 |              |    0 |
|     2 | klasse | VARCHAR(10) |         0 |              |    0 |
|     3 | alter  | INTEGER     |         0 |              |    0 | 

3. Folgende Daten sollen der Tabelle Schueler hinzugefügt werden:

|   id | name   | klasse   | alter   |
|-----:|:-------|:---------|:--------|
|    1 | Lena   | 1B       |         |
|    2 | Peter  | 1B       |         |

4. Setze das Alter von Lena auf 18:

|   id | name   | klasse   |   alter |
|-----:|:-------|:---------|--------:|
|    1 | Lena   | 1B       |      18 |
|    2 | Peter  | 1B       |     nan | 

5. Lösche den Eintrag mit dem Schüler Peter

|   id | name   | klasse   |   alter |
|-----:|:-------|:---------|--------:|
|    1 | Lena   | 1B       |      18 |

6. SQLite macht ohne explizite Angabe bei der Erstellung einer Tabelle keine Type-Checking. Füge einen neuen Schüler mit dem Namen 12345 und dem Alter von "abc" hinzu.

|   id | name   | klasse   | alter   |
|-----:|:-------|:---------|:--------|
|    1 | Lena   | 1B       | 18      |
|    2 | 123456 |          | abc     | 

```python
import sqlite3
import pandas as pd

#1. Tabelle sol immer wieder frisch sein
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

#2. Tabelle erstellen

cursor.execute("""
    CREATE TABLE schueler 
                (sid INTEGER PRIMARY KEY, 
                name VARCHAR(50) NOT NULL,
                klasse VARCHAR(10) NOT NULL,
                'alter' INTEGER NOT NULL);       """)

command = "PRAGMA TABLE_INFO('schueler');"
print(pd.read_sql(command,conn).to_markdown())



cursor.execute("""
INSERT INTO schueler (name,klasse,'alter') VALUES 
            ('Lena','1B',-1),
            ('Peter','1B',-1);
    
""")

pd.read_sql("SELECT * FROM schueler",conn)

#4. Alter von Lena auf 18 setzen

cursor.execute(""" 
UPDATE schueler SET 'alter'=18 WHERE name='Lena';
""")

print(pd.read_sql("SELECT * FROM schueler",conn))


# 5. Lösche den Eintrag mit dem Naem Peter

cursor.execute("""
DELETE FROM schueler WHERE name='Peter';
""")

pd.read_sql("SELECT * FROM schueler",conn)


cursor.execute(""" 
INSERT INTO schueler (name,klasse,"alter") VALUES 
            (12345,TRUE,"abc")
""")
pd.read_sql("SELECT * FROM schueler",conn)
```
